```python
import pandas as pd
```

```python
df = pd.read_csv("minimum_wage_raw.csv")
df.head()
```

```python
data = pd.DataFrame(columns=['year','state','wage'])

for i in range(2012,2022):
    df1 = df[['State',str(i)]]
    df1['year'] = i
    df1 = df1[['year','State',str(i)]]
    df1.columns = ['year','state','wage']
    data = pd.concat([data,df1],axis=0).reindex()
```

```python
data.head()
```

```python
cpi_df = pd.read_csv("cpi.csv")
cpi_df.head()
```

```python
cpi_2021 = float(cpi_df.loc[cpi_df["Year"] == 2021,"CPI"])
cpi_2021
```

```python
data_joined = pd.DataFrame()
data_joined = pd.merge(data,cpi_df,how="left",left_on = ['year'],right_on = ['Year'])
data_joined = data_joined[['year','state','wage','CPI']]
data_joined['CPI_2021'] = cpi_2021
data_joined.head()
```

```python
data_joined['wage'] = data_joined['wage'].str.replace("$","").astype("float")
data_joined.head()
```

```python
data_joined['modified_wage'] = data_joined["wage"] / data_joined["CPI"] * data_joined["CPI_2021"]
data_joined.head()
```

```python
data_joined.to_csv("adjusted_min_wage.csv",index=False)
```
